```python
import numpy as np
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df =pd.read_csv("Algerian_forest_fires_dataset_UPDATE.csv",header=1)
df.head()
# df.shape
df.info()
```

```python
# data cleaning start
```

```python
df.isnull().sum()
df[df.isnull().any(axis=1)]
```

```python
df.columns
```

```python
# df.loc[122]
df.loc[:122,"Region"]=0
df.loc[122:,"Region"]=1
df.info()
```

```python
df[['Region']]=df[['Region']].astype(int)
df.info()
```

```python
# removing null rows from data
df =df.dropna().reset_index(drop=True)
df.isnull().sum()
```

```python
df.loc[122:123]  # this row is also useless because it is a header of nr=ew group table , so i will remove this row
df.iloc[[122]]
```

```python
df=df.drop(122).reset_index(drop=True)
df
```

```python
df.columns
```

```python
# removing spaces from column names 
df.columns=df.columns.str.strip()
df.columns
```

```python
df.head()
# df.info()
```

```python
# changong data type of columns

df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']]=df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']].astype(int)
df.info()
```

```python
objects=[features for features in df.columns if df[features].dtype=='O']
```

```python
for i in objects:
    if i!="Classes":
        df[i]=df[i].astype(float)
```

```python
df.info()
```

```python
# creating nw csv file with cleaned data set
df.to_csv("forest cleaned data.csv",index=False)
df.head()
```
